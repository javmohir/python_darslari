```
db=sql.connect("electricity.db")
cursor=db.cursor()
```

```
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```
cursor.execute("PRAGMA table_info(elect);")
cursor.fetchall()
```

```
pd.read_sql("SELECT * FROM elect LIMIT 15", db, index_col='index')
```

```
query="SELECT DISTINCT country, code FROM elect"
countires=pd.read_sql(query,db)
countires
```

```
"Uzbekistan" in countires['country'].values
```

```
query="SELECT country, coal, gas, hydro, solar, year, oil FROM elect WHERE country='Uzbekistan' AND year>2010"
dfuz=pd.read_sql(query, db)
dfuz
```

```
# BETWEEN
query="SELECT country, coal, gas, hydro, solar, year, oil FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2015"
dfuz=pd.read_sql(query, db)
dfuz
```

```
query="SELECT country, year, coal, gas, hydro, solar, oil FROM elect WHERE country='Uzbekistan' AND year IN (2000,2001,2002)"
dfuz=pd.read_sql(query, db)
dfuz
```

```
query="SELECT country, year, coal, gas, hydro, solar, oil FROM elect WHERE country LIKE '%stan' AND year IN (2019)"
pd.read_sql(query, db)
```

```
query="SELECT country, year,code, coal,hydro,solar,oil FROM elect WHERE code LIKE 'A__' AND year IN (2019)"
pd.read_sql(query, db)
```

```
query="SELECT country,code,coal,hydro,solar,oil FROM elect WHERE code LIKE 'U__' AND year IN (2019)"
pd.read_sql(query, db)
```

```
query="SELECT country, code, hydro, solar, oil, coal FROM elect WHERE code='ALB' AND year BETWEEN 2010 AND 2015"
pd.read_sql(query, db)
```
