```python
db =sql.connect("chinook.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT COUNT (*) FROM customers")
cursor.fetchall()
```

```python
cursor.execute("SELECT COUNT (*) FROM artists")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info (customers);")
cursor.fetchall()
```

```python
cursor.execute("SELECT * FROM customers")
cursor.fetchall()
```

```python
df=pd.read_sql("SELECT * FROM customers LIMIT 10",db, index_col='CustomerId')
df
```

```python
df1=pd.read_sql("SELECT * FROM artists LIMIT 10",db, index_col='ArtistId')
df1
```

```python
query="SELECT CustomerId, FirstName, Country, LastName, Phone FROM customers ORDER BY LastName "
df=pd.read_sql(query,db, index_col='CustomerId')
df.sample(5)
```

```python
query="SELECT CustomerId, FirstName, Country, LastName, Phone From customers ORDER BY Country"
df1=pd.read_sql(query, db, index_col='CustomerId')
df1
```

```python
query="SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
```

```python
pd.read_sql(query, db)
```

```python
query='SELECT * FROM customers WHERE country="USA"'
df2=pd.read_sql(query, db, index_col='CustomerId')
df2.sample(3)
```

```python
query='SELECT * FROM customers WHERE country="USA" OR country="India" ORDER BY City'
pd.read_sql(query, db, index_col='CustomerId')
```

```python
query='SELECT * FROM customers WHERE country="Brazil" OR country="India" ORDER BY LastName'
pd.read_sql(query, db, index_col='CustomerId')
```

```python
query='SELECT * FROM customers WHERE country="USA" AND State="NY" ORDER BY Country'
pd.read_sql(query, db, index_col='CustomerId')
```
